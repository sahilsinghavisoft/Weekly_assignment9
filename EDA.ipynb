```python
date=date_fruit.drop("Class",axis=1)
```

```python
corr = date.corr()
```

```python
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(20, 15))
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'shrink': .7})

plt.show()
```

```python
class_counts = date_fruit['Class'].value_counts()
plt.figure(figsize=(10, 6))  
plt.pie(class_counts, labels=class_counts.index, autopct='%.1f%%')
plt.title('Pie Chart of Date Fruit Classes')  
plt.axis('equal') 
plt.show()
```
